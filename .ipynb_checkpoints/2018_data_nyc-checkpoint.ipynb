```python
import pandas as pd
import os
import csv
```

```python
# import lat/lon csv
all_data = "zipcode_sf_zri_2018.csv"
```

```python
all_data_pd = pd.read_csv(all_data)
all_data_pd.head()
```

```python
# import zip code csv
nyc_zip_codes = "NYC_ZIP.csv"
```

```python
nyc_zip_codes_pd = pd.read_csv(nyc_zip_codes)
nyc_zip_codes_pd.head()
```

```python
# rename column
renamed_nyc_zip_codes = nyc_zip_codes_pd.rename(columns={'10001':'zipcode'})
renamed_nyc_zip_codes.head()
```

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(all_data_pd, renamed_nyc_zip_codes, on="zipcode", how="right")
merge_table.head()
```

```python
merge_table_pd = merge_table.dropna(how='any')
merge_table_pd.head()
```

```python
merge_table.to_csv('nyc_ZRI-201.csv')
```
